```python
from sklearn.datasets import make_classification

x,y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, random_state=1)
```

```python
import numpy as np
np.unique(y)
```

```python
print(len(y[y==0]),len(y[y==1]))
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.5, random_state=1, stratify=y)
```

```python
print(len(ytrain[ytrain==0]),len(ytrain[ytrain==1]))
print(len(ytest[ytest==0]),len(ytest[ytest==1]))
```

### Building Supervised Model ::

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(xtrain, ytrain)
ypreds = model.predict(xtest)
print(accuracy_score(ypreds,ytest)*100)
```

Building Semi-Suipveisd Model ::

```python
xtrain_lab, xtrain_unlab, ytrain_lab, ytrain_unlab = train_test_split(xtrain,ytrain, test_size=0.5, random_state=1, stratify=ytrain)
```

```python
xtrain_mix = np.concatenate((xtrain_lab, xtrain_unlab))
```

```python
len(xtrain_mix)
```

```python
nolab = [-1 for _ in range(len(ytrain_unlab))]
print(len(nolab))
```

```python
ytrain_mix = np.concatenate((ytrain_lab, nolab))
print(len(ytrain_mix))
```

```python
print(len(ytrain_mix[ytrain_mix==0]),len(ytrain_mix[ytrain_mix==1]),len(ytrain_mix[ytrain_mix==-1]))
```

## Label Propagation ::

```python
from sklearn.semi_supervised import LabelPropagation
model_lb = LabelPropagation()
model_lb.fit(xtrain_mix, ytrain_mix)
ypreds_lb = model_lb.predict(xtest)
print(accuracy_score(ypreds_lb, ytest)*100)
```

### Label Spreading

```python
from sklearn.semi_supervised import LabelSpreading
model_ls = LabelSpreading()
model_ls.fit(xtrain_mix, ytrain_mix)
ypreds_ls = model_ls.predict(xtest)
print(accuracy_score(ypreds_ls, ytest)*100)
```
