# SKLearn Methods

```python
import pandas as pd
data = pd.read_csv('/content/sample_data/50_Startups_dataset.csv')
data.head()
```

```python
data.dtypes
```

```python
data['State'] = pd.Categorical(data['State'])
data.dtypes
```

# **Label Encoder :**

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['State']= encoder.fit_transform(data['State'])
data.head()
```

# One-Hot Encoding ::

```python
import pandas as pd
data = pd.read_csv('/content/sample_data/50_Startups_dataset.csv')
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
```

```python
temp = ohe.fit_transform(data['State'].values.reshape(-1,1))
# data.head()
type(temp)
data.head()
```

```python
cats = data['State'].unique()
print(cats)
```

```python
df = pd.DataFrame(temp,columns = cats)
```

```python
df.head()
```

```python
df_new = pd.concat([data,df], axis =1)
df_new.head()
```

```python
df_new.drop(['Unnamed: 0', 'State'], axis =1, inplace = True)
df_new.columns
df_new.head()
```
