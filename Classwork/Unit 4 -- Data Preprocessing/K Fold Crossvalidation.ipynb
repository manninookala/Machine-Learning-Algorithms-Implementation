```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
x = data.data
y = data.target
```

```python
print(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size =0.2, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain,ytrain)
ypreds = model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print('Test Accuracy = ', accuracy_score(ypreds,ytest))
```

```python
train_preds = model.predict(xtrain)
print('Train Accuracy = ', accuracy_score(train_preds, ytrain))
```

## K Fold Cross Validation ::

```python
from sklearn.model_selection import KFold, cross_val_score
kf = KFold(n_splits = 5)
score = cross_val_score(LogisticRegression(),x,y, cv = kf)
```

```python
print(score)
```

```python
score.mean()
```

## Case 2

```python
from sklearn.datasets import load_iris
data = load_iris()
x = data.data
y = data.target
print(x.shape,y.shape)
```

```python
type(y)
```

```python
y[:10]
```

```python
import numpy as np
np.unique(y)
```

```python
# Split
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain,ytest = train_test_split(x,y, test_size=0.3, random_state=1)
print(xtrain.shape, xtest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(xtrain, ytrain)
ypreds = model.predict(xtest)
print(accuracy_score(ypreds,ytest))
```

Apply K-Fold

```python
from sklearn.model_selection import KFold, cross_val_score
kf = KFold(n_splits=5)
score = cross_val_score(LogisticRegression(), x, y,cv=kf)
print(score)
print(score.mean())
```

Apply Stratified for the same dataset.

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits = 5)
score = cross_val_score(LogisticRegression(), x, y,cv=skf)
print(score)
print(score.mean())
```
