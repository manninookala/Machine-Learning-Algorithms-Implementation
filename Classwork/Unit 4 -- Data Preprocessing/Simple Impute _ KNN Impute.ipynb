## **Imputing**

```python
import pandas as pd
data = pd.read_csv('/content/sample_data/Titanic.csv')

data
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

si = SimpleImputer(missing_values=np.nan, strategy = 'mean')
data['Age'] = si.fit_transform(data['Age'].values.reshape(-1,1))

data
```

```python
data['Fare'] = si.fit_transform(data['Fare'].values.reshape(-1,1))
data
```

```python

si = SimpleImputer(missing_values=np.nan, strategy = 'most_frequent')
data['Sex'] =si.fit_transform(data['Sex'].values.reshape(-1,1))
data
```

```python
data.iloc[1,2]=0
data
```

```python
si = SimpleImputer(missing_values=0, strategy = 'mean')
data['Age'] = si.fit_transform(data['Age'].values.reshape(-1,1))
data
```

```python
data.iloc[1,2]= 0
data.iloc[5,2] = -1
data
```

```python
data['Age'].replace([0, -1], np.nan, inplace=True)

# Create a SimpleImputer to fill NaN values with the mean
si = SimpleImputer(strategy='mean')

# Fit and transform the 'Age' column
data['Age'] = si.fit_transform(data['Age'].values.reshape(-1, 1))

data
```

## **KNN Imputer**
  Considers K no. of rows and average the value as imputed value

```python
import pandas as pd
data = pd.read_csv('/content/sample_data/Titanic.csv')

data
```

```python
from sklearn.impute import KNNImputer
ki = KNNImputer(n_neighbors=3)
data['Age'] = ki.fit_transform(data['Age'].values.reshape(-1,1))
data
```

```python
from sklearn.impute import KNNImputer
ki = KNNImputer(n_neighbors=8)
data[['Age','Fare']] = ki.fit_transform(data[['Age','Fare']])
data
```
