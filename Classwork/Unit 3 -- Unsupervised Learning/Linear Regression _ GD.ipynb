```python
# reshape y as 100,1
y= y.reshape((100,1))
print(y.shape)
```

```python
w = np.zeros((x.shape[1],1))
```

```python
def gd(x,y,alpha = 0.001, iters = 100):       #Alpha & iters arethe default parameters.
  w = np.zeros((x.shape[1],1))
  for i in range(iters):
    error =  y - x.dot(w)     # ypred = x*W
    grad = -(x.T).dot(error)
    w = w - alpha * grad
  return w

```

```python
w = gd(x,y)
w
```
