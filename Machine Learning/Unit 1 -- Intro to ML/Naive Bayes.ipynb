```python
#Lets use the inbuilt dataset of sklearn (Wine dataset)
from sklearn.datasets import load_wine
dataset = load_wine()
```

```python
dataset.DESCR
```

```python
dataset.feature_names
```

```python
print(len(dataset.feature_names),dataset.feature_names )
print(dataset.target_names)
```

```python
#Load into x and y
x = dataset.data
y = dataset.target
#x is the faeture matrix & y is the target vector.
```

```python
# Sample values
x[:2]
```

```python
y[-15:]
```

```python
# Because the feature ranges vary, perform Feature scaling.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled[:2]
#All feature values are btw 0-1.
```

```python
#Split X_scaled and y into train and test sets.

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x_scaled,y,test_size= 0.2, random_state=40)
print(xtrain.shape,xtest.shape,ytrain.shape,ytest.shape)
```

```python
# Building the Model

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(xtrain,ytrain)
```

```python
ypreds = model.predict(xtest)
```

```python
ypreds[:5]
```

```python
ytest[:5]
```

```python
# Check the model performance on test dataset.

from sklearn.metrics import confusion_matrix
print(confusion_matrix(ypreds,ytest))     #performance is 100%
```

```python
# Check the model performance on trained dataset.
ypreds_train = model.predict(xtrain)
print(confusion_matrix(ypreds_train,ytrain))

```

```python

from sklearn.metrics import accuracy_score
print('Train Accuracy',accuracy_score(ypreds_train,ytrain)*100)
print('Test Accuracy',accuracy_score(ypreds,ytest)*100)
```
