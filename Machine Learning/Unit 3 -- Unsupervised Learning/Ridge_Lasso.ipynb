```python
from sklearn.datasets import make_regression
```

```python

data = make_regression(n_samples = 100, n_features = 1, n_targets = 1)
data
```

```python
x = data[0]
x[:5]
```

```python
y = data[1]
y[:5]
```

```python
print(x.shape,y.shape)
```

```python
import numpy as np
y = np.reshape(y,(100,1))
y.shape
```

```python
# Ridge regresiion using
from sklearn.linear_model import Ridge
model = Ridge(alpha =1)
```

```python
model.fit(x,y)
print(model.coef_)

```

```python
w = model.coef_
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, color = 'green', label = 'True y')
plt.plot(x, x*w, color = 'red', label = 'Pred y')
plt.legend()
plt.show()
```

```python
model = Ridge(alpha =10)
model.fit(x,y)
print(model.coef_)
w = model.coef_

plt.scatter(x,y, color = 'green', label = 'True y')
plt.plot(x, x*w, color = 'red', label = 'Pred y')
plt.legend()
plt.show()
```

```python
model = Ridge(alpha =15)
model.fit(x,y)
print(model.coef_)
w = model.coef_

plt.scatter(x,y, color = 'green', label = 'True y')
plt.plot(x, x*w, color = 'red', label = 'Pred y')
plt.legend()
plt.show()
```
