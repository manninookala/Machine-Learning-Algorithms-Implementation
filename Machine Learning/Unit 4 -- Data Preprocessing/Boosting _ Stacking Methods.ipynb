## **Boosting:**

```python
import pandas as pd
data = pd.read_csv('/content/sample_data/train.csv')
data.columns
```

```python
data.head()
```

```python
data.info()
```

```python
data = data[['Pclass', 'Age', 'SibSp','Fare', 'Survived']]
data.dropna(inplace = True)
data.info()
```

```python
x = data[[ 'Pclass', 'Age', 'SibSp','Fare']]
y = data['Survived']
```

```python
x[:5]
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.15, random_state=10)
print(xtrain.shape,xtest.shape,ytrain.shape,ytest.shape)
```

## **AdaBoost Method : **

```python
from sklearn.ensemble import AdaBoostClassifier,  GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

abc = AdaBoostClassifier(n_estimators = 20,base_estimator=LogisticRegression(), learning_rate=1)
abc.fit(xtrain, ytrain)
ypreds = abc.predict(xtest)
print(accuracy_score(ypreds,ytest))
```

## **Gradient Boost : **

```python
gbc = GradientBoostingClassifier(n_estimators = 20,max_depth=1, learning_rate=1)      #GradientBoostClassifier doesnot take base_estimator as it uses DT by default.
gbc.fit(xtrain, ytrain)
ypreds = gbc.predict(xtest)
print(accuracy_score(ypreds,ytest))
```

## **Stacking :**

```python
from sklearn.datasets import make_classification
x, y = make_classification(n_samples=1000, n_features=5, random_state =1)
```

```python
print(type(x), type(y))
```

```python
x[:5]
```

```python
y[:5]
```

```python
import numpy as np
np.unique(y)            # Binary Classification dataset with 5 features
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

level0 = [
    ('lr', LogisticRegression()),
    ('dt', DecisionTreeClassifier()),
    ('svc', SVC())
]

level1 = LogisticRegression()
```

```python
from sklearn.ensemble import StackingClassifier
model = StackingClassifier(estimators=level0, final_estimator=level1)
model.fit(xtrain,ytrain)
```

```python
ypreds = model.predict(xtest)
from sklearn.metrics import accuracy_score
print(accuracy_score(ypreds, ytest))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

level0 = [
    ('lr', LogisticRegression()),
    ('dt', DecisionTreeClassifier()),
    ('svc', SVC())
]

# level1 = [
#     ('lr1', LogisticRegression()),
#     ('nb', GaussianNB())
# ]
level2 = DecisionTreeClassifier()


from sklearn.ensemble import StackingClassifier
model = StackingClassifier(estimators=level0, final_estimator=level2)
model.fit(xtrain,ytrain)

ypreds = model.predict(xtest)
from sklearn.metrics import accuracy_score
print(accuracy_score(ypreds, ytest))
```
