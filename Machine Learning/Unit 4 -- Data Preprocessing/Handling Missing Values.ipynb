```python
# Use regression model to predict the Null values.

data.columns
```

```python
# Fare of ticket depends on Age and Class features.

data_new = data[['Age', 'Pclass', 'Fare']]
```

```python
#Take Age and Pclass as features and Fare as target.

data_new
```

```python
data_new['Age'].fillna(data['Age'].mean(), inplace = True)
data_new
```

```python
test = data_new.iloc[8,:]
```

```python
train = data_new.drop(index = 8, axis = 0)
```

```python
train
```

```python
trainx = train[['Age','Pclass']]
trainy = train['Fare']

trainx
```

```python
trainy
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(trainx,trainy)
```

```python
test
```

```python
testx = test[['Age','Pclass']]
```

```python
pred_fare = lr.predict([testx])
pred_fare
```

```python
pred_fare = lr.predict(testx.values.reshape((1,-1)))
pred_fare
```
