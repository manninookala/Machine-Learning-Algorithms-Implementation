Most of the ML Algorithms cannot process data other than numerics. Hence categorical  columns have to be converted to numerics called **Data Encoding**.

  Data Encoding techniques are ::

    In sklearn ::
      1. Lable encoder
      2. One-hot encoding
    In Pandas ::
      3. Replace method
      4. Get_dummies

  Categorical feartures are of 2 typres ::

  1. Nominal - values have no order. Ex, Gender
  2. Ordinal - values halues have some order. Ex, shirt sizes.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df.head()
```

```python
df['Embarked'] = le.fit_transform(df['Embarked'])
df.head()
```

```python
# As we are going to  use clustering Algorithm remove the Target(Survived).
df.drop('Survived', axis = 'columns', inplace = True)
df.head()
```

```python
# Normalise the data
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
data_scaled = mms.fit_transform(df)
data_scaled[:5]
```

```python
#Perform Clustering using K-means.

from sklearn.cluster import KMeans
modal = KMeans(n_clusters = 3)
modal.fit(data_scaled)
print(modal.inertia_)
```

```python
performance =[]
for i in range(2, 50):

  modal = KMeans(n_clusters = i, init = 'k-means++')
  modal.fit(data_scaled)
  performance.append(modal.inertia_)
```

```python
performance
```

```python
import matplotlib.pyplot as plt
plt.plot(range(2,50), performance)
plt.show()
```

```python
# How to chooose the K-value : Elbow method.
# Draw graph of Algorithm performance& choose that k where there performance stabilise or error starts incresing.
```

Initialising the centroids can be chosen using ::
  1. Randomly select the centroids
  2. k-means++

```python
performance =[]
for i in range(2, 50):

  modal = KMeans(n_clusters = i, init = 'random')
  modal.fit(data_scaled)
  performance.append(modal.inertia_)

plt.plot(range(2,50), performance)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
model = AgglomerativeClustering(n_clusters = 5)
model.fit(data_scaled)
print(model.silhouette_score)
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN()
model.fit(data_scaled)
print(model.inertia_)
```
