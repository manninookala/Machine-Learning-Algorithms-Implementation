# **K Means Clustering ::**


Clustering is given datapoints, divide the datapoints into groups called clusters.

It is an **unsupervised learning method**.

We consider only feature matrix 'x' and no target vector.

**K-Means** is the Clustering algorithms based on Distance measure to form the clusters.

It uses Euclidian distance.

  Datapoint is a tupple with all feature values.
  
Every cluster will have cluster header also called as **Cluster Centroid**.

Procedure Steps ::
  1.   Arbitarily take 'n' centroids and calculate the distance of each centroid from every datapoint.
  2.   Consider the least distance and assign it to that centroid.
  3. Calculate the

**|Algorithms ::**
  1. Choose the k value(no. of clusters).
  2. Choose initial K no. of centroids.
  3. Repeat ;
        1. Calculate the distance for each datapoint to k centroids.
        2. Assign the datapoint to that cluster that is closest.
        3. Calculate the average distance of the datapoints belonging to each cluster.
        4. Do the (3) for all the clusters.
        5. These average distance become new centroinds.

    Until convergence.

**|Convergence Criteria ::**
  1. Until datapoints do not change the clusters.
  2. After spcified no. of iterations.
  3. Centroids are not changing.

```python
# Form the clustering
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
km.fit(data)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
km.fit(data)
km.inertia_
```

```python
#Increasing the Clusters to decrease the inertia
for i in range(2,25):

  km = KMeans(n_clusters=i)
  km.fit(data)
  print(km.inertia_)
```
